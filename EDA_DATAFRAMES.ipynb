```python

def summarize(df: pd.DataFrame, name: str):
    print(f"\n=== Resumen de {name} ===")
    print(f"- Filas   : {len(df)}")
    print(f"- Columnas: {list(df.columns)}")
    print("\nPrimeros 5 registros:")
    print(df.head(5))
    # Longitud de texto
    if "text" in df.columns:
        print(f"\nLongitud media de texto: {df['text'].str.len().mean():.1f} caracteres")
    elif "texto" in df.columns:
        print(f"\nLongitud media de texto: {df['texto'].str.len().mean():.1f} caracteres")

    # Si hay columna 'sentiment' (Sentiment140 raw)
    if "sentiment" in df.columns:
        counts = df["sentiment"].value_counts().sort_index()
        print("\nDistribución de 'sentiment':")
        print(counts)


def main():
    # 1) Carga los dos CSV
    tweets = pd.read_csv(
        "training.1600000.processed.noemoticon.csv",
        encoding="latin-1",
        names=["sentiment","id","date","query","user","text"],
        header=None
    )
    support = pd.read_csv(
        "twcs.csv",
        encoding="utf-8"
    )

    # 2) Resúmenes
    summarize(tweets,  "Sentiment140 Tweets")
    summarize(support, "Tickets de Soporte")

if __name__ == "__main__":
    main()
```
